# Map of UK postcodes obtained with Postcodes_grid.py

```python
# Import libraries
import folium as fm
import pandas as pd
import geopandas as gpd
```

## Latitude and longitude from the GEODATA grid

```python
# Import the data
grid=gpd.read_file("grid_10km_point.gpkg")
grid_UK=grid[grid["CNTR_ID"]=="UK"]
grid_projected=grid_UK.to_crs(epsg=4326)
grid_projected['lon'] = grid_projected['geometry'].x
grid_projected['lat'] = grid_projected['geometry'].y
```

```python
# Create a map of the UK marking the locations of the grid points as red circles
map_pstcd = fm.Map( location=[54.673511, -5.999533], zoom_start=6.)
for index in range(len(grid_projected)):
    fm.CircleMarker(radius=1,location=[grid_projected.lat.iloc[index], grid_projected.lon.iloc[index]],  color='crimson',
    fill=True,
).add_to(map_pstcd)
```

## Latitude and longitude of the postcodes obtained with Postcode_grid.py

```python
# Import the postcode data
Postcodes=pd.read_csv('Postcodes_grid.csv')
Postcodes.head()
```

```python
# Marks as black circles the points for which postcodes were obtained
for index in range(len(Postcodes)):
    fm.CircleMarker(radius=5,location=[Postcodes.latitude[index], Postcodes.longitude[index]],  popup= Postcodes.Postcode[index], color='black',fill=True).add_to(map_pstcd)
```
